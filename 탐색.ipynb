#### **예제 1 : 순차 탐색**

- 정렬되지 않은 리스트에서도 원하는 레코드를 찾을 수 있는 가장 단순하고 직관적인 방법
- 리스트의 각 데이터를 처음부터 하났기 순서대로 검사하여 원하는 데이터를 탐색
- 효율적이지 않음

```python
# 순차 탐색(sequential search)

data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
key = 8

for i in range(len(data)): 
    if data[i] == key:   #탐색이 성공하면 
        print(i)         #인덱스를 반환
```

#### **예제 2 : 이진 탐색**

- 리스트의 중앙에 있는 값을 조사하여 찾는 항목이 왼쪽에 있는지 오른쪽에 있는지 판단
- 찾는 항목이 왼쪽에 있다면 오른쪽은 탐색할 필요가 없음
- 매 단계마다 검색해야 할 항목수가 반으로 줄어듬

```python
# 이진탐색(binary search)

data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
key = 2
low = 0
high = len(data)

while low < high:
    middle = (low+high)//2

    if key == data[middle]:
        break
    elif key > data[middle]:
        low = middle + 1
    else:
        high = middle - 1
    
if data[middle] == key:
    print(f'{key}는 {middle} 위치에 있습니다.')
else:
    print(f'{key}를 찾을 수 없습니다.')
```
